```python
# Exploring Genetic Drift with Python: A Random Walk Simulation

import matplotlib.pyplot as plt
import seaborn as sns
import random

# Apply Seaborn's default styling
sns.set()

class GeneticDrift:
    def __init__(self, num_points=100):
        self.num_points = num_points
        self.allele_frequencies = [0.5]  # Start with a 50% allele frequency

    def simulate_drift(self):
        while len(self.allele_frequencies) < self.num_points:
            current_frequency = self.allele_frequencies[-1]
            change = random.choice([-0.01, 0.01])  # Small random change
            new_frequency = current_frequency + change
            
            # Ensure frequency stays between 0 and 1
            if new_frequency < 0:
                new_frequency = 0
            elif new_frequency > 1:
                new_frequency = 1

            self.allele_frequencies.append(new_frequency)

# simulating the genetic drift and visualizing it using Matplotlib.

# Simulate genetic drift
gd = GeneticDrift(num_points=500)
gd.simulate_drift()

# Plot the allele frequencies over time
plt.figure(figsize=(10, 6))
time_points = range(gd.num_points)
plt.plot(time_points, gd.allele_frequencies, c='blue')

# Highlight the starting and ending points
plt.scatter(0, gd.allele_frequencies[0], c='green', edgecolor='none', s=100, label='Start')
plt.scatter(gd.num_points-1, gd.allele_frequencies[-1], c='red', edgecolor='none', s=100, label='End')

# Set chart title and labels
plt.title("Genetic Drift Simulation", fontsize=24)
plt.xlabel("Generation", fontsize=14)
plt.ylabel("Allele Frequency", fontsize=14)
plt.legend()

# Show the plot
plt.show()
```
